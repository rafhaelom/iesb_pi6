```python
def conecta_banco(host, port, dbname, user, password):
    """
    Função para criar tabela no banco
    """
    # Conexão ao banco de dados.
    conn = psycopg2.connect(host=host, port=port, dbname=dbname, user=user, password=password)
    return conn

def consultar_banco(sql, colunas):
    """
    Função para realizar consultas no banco de dados.
    """
    registros = []
    # Conecta ao banco de dados.
    con = conecta_banco(host, port, dbname, user, password)
    cur = con.cursor()
    cur.execute(sql)
    recset = cur.fetchall()
    
    for rec in recset:
        registros.append(rec)
    con.close()
    
    # Tranformando os dados da consulta no PostegreSQL em DataFrame
    df = pd.DataFrame(registros, columns=colunas)
    return df
```

```python
def SIHSUS_MUN_ALL(ano, colunas_sih_mun):
    """
    Função para retornar dataframe com todos os codigos de municipio por ano.
    """
    if ano != 2022 and ano <=2021:
        # Mês 1 = Janeiro.
        df_1 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 1) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 2 = Fevereiro.
        df_2 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 2) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 3 = Março.
        df_3 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 3) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 4 = Abril.
        df_4 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 4) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 5 = Maio.
        df_5 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 5) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 6 = Junho.
        df_6 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 6) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 7 = Julho.
        df_7= consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 7) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 8 = Agosto.
        df_8 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 8) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 9 = Setembro.
        df_9 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 9) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 10 = Outubro.
        df_10 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 10) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 11 = Novembro.
        df_11 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 11) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 12 = Dezembro.
        df_12 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 12) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Cria lista de dataframes.
        frames = [df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11, df_12]

        # Concatena dataframes 'df_ano'.
        df_ano = pd.concat(frames, axis=0, keys=[f"{ano}_1", f"{ano}_2", f"{ano}_3", f"{ano}_4", f"{ano}_5", \
                                                 f"{ano}_6", f"{ano}_7", f"{ano}_8", f"{ano}_9", f"{ano}_10", \
                                                 f"{ano}_11", f"{ano}_12"])

        # Reset do index dataframe.
        df_ano.reset_index(drop=False, inplace=True)

        # Dropa colunas que não serão utilizadas.
        df_ano.drop(columns=["level_1", "ID_SIH_QT", "ANO", "MES", "COD_MUNICIPIO"], axis=1, inplace=True)

        # Renomeia colunas.
        df_ano.rename(columns={"level_0": "ANO_MES", "COD_COMPLETO_MUNICIPIO_SEM_DV": "COD_MUNICIPIO_SEM_DV"}, inplace=True)

        # Cria coluna ano e mes.
        df_ano.insert(1, "ANO", df_ano.ANO_MES.apply(lambda x: x[:4]), allow_duplicates=False)
        df_ano.insert(2, "MES", df_ano.ANO_MES.apply(lambda x: x[5:]), allow_duplicates=False)
        return df_ano
    else:
        # Mês 1 = Janeiro.
        df_1 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 1) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 2 = Fevereiro.
        df_2 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 2) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 3 = Março.
        df_3 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 3) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)

        # Mês 4 = Abril.
        df_4 = consultar_banco(sql=f'''
                                    select 
                                        a."COD_COMPLETO_MUNICIPIO_SEM_DV",
                                        b.*
                                    from sih_sus.tb_dtb_municipios_ibge a
                                        left join (select * from sih_sus.tb_sus_sih_qt b
                                                        where b."ANO" = {ano} and b."MES" = 4) b
                                            on a."COD_COMPLETO_MUNICIPIO_SEM_DV" = b."COD_MUNICIPIO";''', \
                             colunas=colunas_sih_mun)
        
        # Cria lista de dataframes.
        frames = [df_1, df_2, df_3, df_4]

        # Concatena dataframes 'df_ano'.
        df_ano = pd.concat(frames, axis=0, keys=[f"{ano}_1", f"{ano}_2", f"{ano}_3", f"{ano}_4"])

        # Reset do index dataframe.
        df_ano.reset_index(drop=False, inplace=True)

        # Dropa colunas que não serão utilizadas.
        df_ano.drop(columns=["level_1", "ID_SIH_QT", "ANO", "MES", "COD_MUNICIPIO"], axis=1, inplace=True)

        # Renomeia colunas.
        df_ano.rename(columns={"level_0": "ANO_MES", "COD_COMPLETO_MUNICIPIO_SEM_DV": "COD_MUNICIPIO_SEM_DV"}, inplace=True)

        # Cria coluna ano e mes.
        df_ano.insert(1, "ANO", df_ano.ANO_MES.apply(lambda x: x[:4]), allow_duplicates=False)
        df_ano.insert(2, "MES", df_ano.ANO_MES.apply(lambda x: x[5:]), allow_duplicates=False)
        return df_ano
```
